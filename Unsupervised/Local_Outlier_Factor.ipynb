#DB-Scan Clustering with Local Outlier Factor

#Importing independencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples = 750, factor = 0.25, noise = 0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps = 0.10)
```

```python
db.fit(X)
```

```python
db.fit_predict(X)
```

```python
db.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.scatter(X[:,0],X[:,1], c=db.labels_)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=20,algorithm='auto')
```

```python
lof.fit(X)
```

```python
lof.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=lof.fit_predict(X))
```

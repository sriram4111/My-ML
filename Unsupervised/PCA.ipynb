#PCA(Principal Component Analysis) Implementation

#Importing independencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Load dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_df = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer_df['data'],columns=cancer_df['feature_names'])
```

```python
df.head()
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data #--> #these are my features
```

```python
#Apply PCA AlgorithmS
from sklearn.decomposition import PCA
```

```python
pca  = PCA(n_components=2)
```

```python
pca_data =  pca.fit_transform(scaled_data)
```

```python
pca_data #--> #Data Coverted into 2 features
```

#It transform dataset into 2 features this we call as Feature Extraction
#it coverted in a way that Max Variance captured..!

```python
pca.explained_variance_
```

#This is my variance of feature 1 and 2 from PCA Feature Extracion

```python
plt.figure(figsize=(10,6))
plt.scatter(x=pca_data[:,0],y=pca_data[:,1],c=cancer_df['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

plt.scatter(x=pca_data[:,0], y=pca_data[:,1], c=cancer_df['target']): This is the core of the plotting.  It creates a scatter plot, where each point represents a data sample.

pca_data[:,0]:  pca_data is assumed to be a NumPy array (or something similar) containing the data after PCA has been applied.  [:,0] selects all rows (:) and the first column (0).  This means it's taking the values of the first principal component for all data points.  These values will be used for the x-coordinates of the scatter plot.

pca_data[:,1]: Similar to the above, this selects all rows and the second column (1) from pca_data.  These are the values of the second principal component, and they'll be used for the y-coordinates of the scatter plot.

c=cancer_df['target']: This controls the color of each point in the scatter plot.  cancer_df is likely a Pandas DataFrame containing information about the data, and 'target' is the name of a column in that DataFrame.  This column probably represents some categorical variable (e.g., whether a patient has cancer or not).  The c argument maps the values in the 'target' column to different colors, so points with the same target value will have the same color, making it easy to visually distinguish groups in the PCA-transformed data.

```python
#Import independencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Quadratic Equation - ax^2+bx+c = 0  ---> 0.5*X^2+1.5X+2+outliers

```python
#creating a nonlinear dataset of independent feature and dependent feature by using numpy
X = 6 * np.random.rand(100,1)-3
Y = 0.5*X**2 + 1.5*X + 2 +np.random.randn(100,1)
```

#### Here you can see the pattern it follows the Non-Linear RelationShip

```python
 ## Plotting my dataset
plt.scatter(X,Y,color = 'green')
plt.title("DataSet")
plt.xlabel('X_Data')
plt.ylabel('Y_Data')
```

**We try to consider this DataSet and we try to Implement linear regression and ploynomial regression we try to see how the accuracy is going to improved.**

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

Let's Implement **Simple Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression =LinearRegression()
```

**We are not going to perform DataPreprocessing (or) Standardization Because X,Y features are already in continous variables form.**

```python
regression.fit(X_train,Y_train)
```

**Checking performance of my model by using r2_score**

```python
from sklearn.metrics import r2_score
```

**r2_score(Y_test, regression.predict(X_test)):**

r2_score is a metric from sklearn.metrics used to evaluate the accuracy of a regression model.
Y_test is the actual (true) values of the target variable for the test set.
regression.predict(X_test) uses the trained regression model to predict target values for the test set X_test

```python
score = r2_score(Y_test,regression.predict(X_test))
```

```python
score
```

Let's Visualize this model

```python
plt.scatter(X_train,Y_train,color='orange')
plt.scatter(X_train,regression.predict(X_train),color='red')
```

**This line uses the trained regression model to predict target values based on X_train.,**
**regression.predict(X_train) outputs the predicted values for each data point in X_train, which are then plotted against X_train itself.**
**When you get this best fit line you will get Good amount of Error High.**
Let's see how we can slove this problem with **Ploynomial Regression**

How we can Create Degree for 2,3,4
Let's apply polynomial transformation
In sklearn we have PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly =  poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
#linear regression again
from  sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
regression_1 = LinearRegression()
```

```python
regression_1.fit(X_train_poly,Y_train)
```

```python
y_pred = regression_1.predict(X_test_poly)
print(y_pred)
```

```python
score = r2_score(y_pred,Y_test)
print(score)
```

```python
print(regression_1.coef_)
```

```python
print(regression_1.intercept_)
```

```python
#best fit line
plt.scatter(X_train,Y_train)
plt.scatter(X_train,regression_1.predict(X_train_poly),color='red')
```

##Increasing Degree=3

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly =  poly.transform(X_test)
```

```python
X_train_poly
```

```python
from  sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
regression_1 = LinearRegression()
```

```python
regression_1.fit(X_train_poly,Y_train)
```

```python
y_pred = regression_1.predict(X_test_poly)
print(y_pred)
```

##Model performance has increased 1.6% 84% accuracy now, Previously when degree=2 we get 82% accuracy

```python
score = r2_score(y_pred,Y_test)
print(score)
```

**####Prediction for newdataset**

```python
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
X_new_poly
```

```python
##predicting
y_new = regression_1.predict(X_new_poly)
plt.plot(X_new,y_new,"r-",linewidth=2,label="Predictions")
plt.plot(X_train,Y_train,"b.",label="Training Points")
plt.plot(X_test,Y_test,"g.",label="Testing Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

#Importing independencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("/content/ttt.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] =df['Gender'].replace('Fe Male','Female')
```

```python
df['Gender'].value_counts()
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['MaritalStatus'] =df['MaritalStatus'].replace('Single','Unmarried')
```

```python
df['MaritalStatus'].value_counts()
```

```python
df.columns
```

```python
features_na = [feature for feature in df.columns if df[feature].isnull().sum()>1]
for i in features_na:
  print(i,np.round(df[i].isnull().mean()*100,5),"%Missing_Values")
```

```python
df
```

```python
df.Age.fillna(df['Age'].median(), inplace=True)
```

```python
df.DurationOfPitch.fillna(df['DurationOfPitch'].median(),inplace=True)
```

```python
df.TypeofContact.fillna(df['TypeofContact'].mode()[0],inplace=True)
```

```python
df.NumberOfFollowups.fillna(df['NumberOfFollowups'].mode()[0],inplace=True)
```

```python
df.MonthlyIncome.fillna(df['MonthlyIncome'].median(),inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python

#get all numerics features and store in one variable
num_features = [features for features in df.columns if df[features].dtype !='O']
print("Number of Numerical Features is:",len(num_features))
```

```python
cat_features = [features for features in df.columns if df[features].dtype =='O']
print("Number of Categorial Features is:", len(cat_features))
```

```python
X = df.drop(['ProdTaken'],axis=1)
Y = df['ProdTaken']
```

```python

#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=10)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python

#Create Column Transformers with 3 types of transformers
cat_features = X.select_dtypes(include='object').columns
num_features = X.select_dtypes(exclude='object').columns
```

```python

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python

num_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')
```

```python
preprocesser = ColumnTransformer([

                              ("OneHotEncoder",oh_transformer,cat_features),
                              ("num_transformer",num_transformer,num_features)
]
)
```

```python

X_train = preprocesser.fit_transform(X_train)
```

```python

pd.DataFrame(X_train)
```

```python
X_test = preprocesser.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score,precision_score,recall_score,roc_auc_score
```

```python
models ={

           "RandomForestClassifier":RandomForestClassifier(),
           "LogisticRegression":LogisticRegression(),
           "KNeighborsClassifier":KNeighborsClassifier(),
           "SVC":SVC(),
           "GaussianNB":GaussianNB(),
           "GradientBoostingClassifier":GradientBoostingClassifier(),
            "DecisionTreeClassifier":DecisionTreeClassifier(),
}

```

```python
for i in range(len(list(models))):
  model = list(models.values())[i]
```

```python
model.fit(X_train,Y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
print(list(models.keys())[i])
```

```python
acc_score  = accuracy_score(Y_train,y_train_pred)
roc = roc_auc_score(Y_train,y_train_pred)

test_acc  = accuracy_score(Y_test,y_test_pred)
roc = roc_auc_score(Y_test,y_test_pred)
```

```python
test_acc  = accuracy_score(Y_test,y_test_pred)
roc = roc_auc_score(Y_test,y_test_pred)
```

```python
#Print Model Performance Training Set
print("Model Performance for Training Set")
print("Accuracy {:.4f}".format(acc_score))
print("ROC_AUC_Score {:.4f}".format(roc))

print("\n-----------------------------------------------\n")

#Print Model Performance  Testing Set
print("Model Performance for Testing Set")
print("Accuracy {:.4f}".format(test_acc))
print("ROC_AUC_Score {:.4f}".format(roc))
```

#Hyperparameter Tuning

```python
Gb_params = {
              "loss":['log_loss','deviance','exponential'],
              "criterion":['friedman_mse','mse','mae'],
              "min_samples_split":[2,8,15,20],
              "max_depth":[5,8,15,None,10],
              "n_estimators":[100,200,500]
}
dt_params = {

             "Criterion":['gini', 'entropy', 'log_loss'],
             "min_samples_split":[2,5,8,10],
             "max_depth":[2,None,5,10],
             "min_samples_leaf":[1,2,3]
}
```

```python
Gb_models = [

             ("GB",GradientBoostingClassifier(),Gb_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model_name = {}
for name,model,params in Gb_models:
    rand = RandomizedSearchCV(estimator=model,param_distributions=params,cv=3,verbose=2,n_jobs=-1)

```

```python
rand.fit(X_train,Y_train)
```

```python
model_name[name] = rand.best_params_
```

```python
for mode in model_name:
  print(f"Best parameters for {mode}")
  print(model_name[mode])
```

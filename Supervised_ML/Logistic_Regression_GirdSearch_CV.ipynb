#Logistic Regression Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#It will create the dataset in classification approach it will create independent feature it will able to create dependent feature in output we can say that how many outputs we want, It will create dataset in such a way that we don't even have to do Standardiztion.

```python
#it will create Dataset
from sklearn.datasets import make_classification
```

```python
#creating Dataset
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2)
```

```python
X
```

```python
y
```

```python
#Converting X into Dataframe
df = pd.DataFrame(X)
df
```

```python
##Converting y into Dataframe
y_df = pd.DataFrame(y)
y_df
```

##Train-Test-Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape,X_test.shape
```

#Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
regression
```

```python
#model training
regression.fit(X_train,y_train)
```

```python
#predict new data
y_pred = regression.predict(X_test)
y_pred
```

#Performance metrics in Logistic Regression

#Accuracy
#Confusion Matrix
#Precision
#Recall

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
```

```python
confusion = confusion_matrix(y_test,y_pred)
print(confusion)
class_report = classification_report(y_test,y_pred)
print(class_report)
```

#Hyperparameter Tunning and Cross-Validation


```python
model = LogisticRegression()
model
```

```python
penalty = ['l1','l2','elastic_net']
c_values = [100,10,1.0,0.1,0.01]
solver = ['netwon_cg','lbfgs','liblinear','saga','sag']
```

```python
#converting list data into key:value_pairs(Dict)
params = dict(penalty=penalty,C=c_values,solver=solver)
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

#For Hyperparameter we use **GirdSearch CV**

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
grid
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
confusion = confusion_matrix(y_test,y_pred)
print(confusion)
class_report = classification_report(y_test,y_pred)
print(class_report)
```

```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read the pd file
df = pd.read_csv("/content/Sonar data.csv",header=None)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Separate the vaues for independent and dependent features
X = df.drop([60],axis=1)
X
```

```python
y = df[60]
y
```

```python
# i will check how many Rock and Mines
df[60].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
```

```python
model   = LogisticRegression()
```

```python
#train the model of giving data
model.fit(X_train,y_train)
```

```python
#metrics
from sklearn.metrics import accuracy_score
```

```python
#We can check the accuracy for training data
pred_X_train = model.predict(X_train)
train_score =  accuracy_score(pred_X_train,y_train)
train_score
```

```python
#We can check the accuracy for Testing data
pred_X_test = model.predict(X_test)
test_score =  accuracy_score(pred_X_test,y_test)
test_score
```

```python
#predict the data
y_pred = model.predict(X_test)
```
